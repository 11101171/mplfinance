```python
def rcParams_to_df(rcp):
    keys = []
    vals = []
    for item in rcp:
        keys.append(item)
        vals.append(rcp[item])
    df = pd.DataFrame(vals,index=pd.Index(keys,name='rcParams Key'))    
    df.columns = ['Value']
    return df
```

```python
df = rcParams_to_df(mpl.rcParams)
```
