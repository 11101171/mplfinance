# Create a Renko plot

### You can specify a few different arguments in `renko_params`:

- **`brick_size`**
    - Can be either an integer/float or 'atr'
        - By specifying an integer or float you are setting each brick to be that value
        - Setting `brick_size='atr'` means the `brick_size` will be set to the Average True Value over the specified `atr_length`
    - `brick_size=2` means for a renko brick to be created the price must move at least a value of 2.
    - Default if no `brick_size` is specified is 2.0
- **`atr_length`**
    - Integer representing the time period over which to calculate the Average True Value
    - If `brick_size` is not 'atr' then setting `atr_length` has no effect on the plot
    - Default if no `atr_length` is specified is 14

Note: `renko_params` must be passed in as a `dict`

---

### We can create a simple renko chart by setting the `type='renko'`

```python
mpf.plot(daily,type='renko')
```

---

### We can adjust the brick size by specifying integer or float values
- Notice how the distance between each date varies

```python
mpf.plot(daily,type='renko',renko_params=dict(brick_size=7))
mpf.plot(daily,type='renko',renko_params=dict(brick_size=2.5))
```

---

### We can also set the brick size to be 'atr' to set the size of each brick to the Average True Value

```python
mpf.plot(daily,type='renko',renko_params=dict(brick_size='atr'))
```

---

### By default the atr_length is 14, but we can change this by specifying `atr_length`

```python
mpf.plot(daily,type='renko',renko_params=dict(brick_size='atr', atr_length=5))
```

# Using other plot functionalities with Renko charts

- All regular `plot()` arguments are valid for Renko charts
    - *Note*: Data you add with `addplot` may look different than what you expect as Renko charts often do not evenly distribute dates over the x-axis
        - When given a set of y values each corresponding to its own unique date, when put on the Renko plot, y values are handled in a specific way to account for the irregular x-axis
            - When there is a set of missing dates in a row in the data, y values are averaged over missing dates and the next available date to find the y value of the next available date
            - When there are duplicate dates in the Renko chart's x values, the price is duplicated for all dates over consecutive x values with the same date
- It is also possible to customize the colors of the Renko plot.

***
### Missing date example with addplot/volume data
- ex. Given a dataset with (date, price) pairs where:
    - *Dataset* = (1, 40.3), (2, 39.2), (3, 40.7), (4, 40.1)
    - In this example, one Renko brick is created on date=1 and one Renko brick is created on date=4.  The data set now looks like:
    - *Data to plot on the Renko chart* = (1, 40.3), (4, 40)
    - This is because the y values: 39.2, 40.7, and 40.1 are averaged: (39.2+40.7+40.1)/3 = 40 leading to the adjusted date=4 pair of (4, 40)
### Duplicate date example with addplot/volume data
- ex. Given a dataset with (date, price) pairs where:
    - Dataset = (1, 40.3), (2, 39.2), (3, 40.7), (4, 40.1)
    - In this example, one Renko brick is created for date = 1, 2, 4, but two Renko bricks are created for date=3
    - *Data to plot on the Renko chart* = (1, 40.3), (2, 39.2), (3, 40.7), (4, 40.7), (5, 40.1)
    - Since there were two Renko bricks on date=3, the price value on that x value was duplicated on date=4 and the succeeding data is shifted

```python
# First we set the kwargs that we will use for the style examples:
kwargs = dict(type='renko',mav=(4,6,8),renko_params=dict(brick_size=7), volume=True,figratio=(10,8),figscale=0.75)
```

```python
mpf.plot(daily,**kwargs,style='blueskies')
```

```python
mpf.plot(daily,**kwargs,style='starsandstripes')
```

---

- within `marketcolors` one can specify colors for the `edges` and `up` and `down` brick colors.

for example:

```python
mc = mpf.make_marketcolors(up='g',down='r',
                           edge='black',
                           volume='gray')
s  = mpf.make_mpf_style(marketcolors=mc)

mpf.plot(daily,**kwargs,style=s)
```
